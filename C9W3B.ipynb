# Segmenting and Clustering Neighborhoods in Toronto II

#### Step 1. Import libraries.

```python
from pandas.io.html import read_html as pdhtml
import pandas as pd
```

#### Step 2. Import and clean dataset

```python
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfhtml = pdhtml(page, index_col=None, header=0, attrs={"class":"wikitable"})
df = dfhtml[0]
df = df.drop(df[df.Borough == 'Not assigned'].index)
df = df.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df.at[85, 'Neighbourhood'] = 'Queen\'s Park'
df.head()
```

#### Step 3. Add Latitude and Longitude within above dataset .

```python
df_geodata = pd.read_csv('https://cocl.us/Geospatial_data')
df_geodata.head()
```

```python
for indice_fila, fila in df.iterrows():
    for indice_fila2, fila2 in df_geodata.iterrows():
        if fila2['Postal Code'] == fila['Postcode']:
            df.at[indice_fila,'Latitude'] = fila2['Latitude']
            df.at[indice_fila,'Longitude'] = fila2['Longitude']
```
